```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('dataset.csv')
dataset.keys()
```

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
x = np.asarray(dataset)
print(x)
```

```python
#Menampilkan data dalam bentuk scatter plot
plt.scatter(x[:,0], x[:,1],label='True Position') 
plt.xlabel("Luas Lahan")
plt.ylabel("Hasil Produksi")
plt.title("Grafik Perkiraan Hasil produksi")
plt.show()
```

```python
#Mengaktifkan K-Means dengan jumlah K=3
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
```

```python
#Menampilkan nilai Centroid yang digenerate oleh algoritma
print(kmeans.cluster_centers_)
```

```python
#Plot Data Point
#Memvisualisasikan Hasil Klasterisasi Data Konsumen
plt.scatter(x[:,0], x[:,1], c=kmeans.labels_,cmap='rainbow')
plt.xlabel("Luas Lahan")
plt.ylabel("Hasil Produksi")
plt.title("Grafik Perkiraan Hasil produksi")
plt.show()
```

```python
#Plot Data Point
#Memvisualisasikan hasil Klasterisasi dengan cetroiud dr masing2 klaster
plt.scatter(x[:,0], x[:,1], c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel("Luas Lahan")
plt.ylabel("Hasil Produksi")
plt.title("Grafik Perkiraan Hasil produksi")
plt.show()
```
